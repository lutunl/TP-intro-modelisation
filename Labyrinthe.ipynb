## Labyrinthe

```python
# Prérequis
import tkinter as tk
import random
```

### Création d'un labyrinthe

```python
# Paramètres
LARGEUR = 21
HAUTEUR = 21
TAILLE_CASE = 25

# Création d'une grille remplie de murs
maze = [[1 for _ in range(LARGEUR)] for _ in range(HAUTEUR)]

# Génération du labyrinthe
def generate(x, y):
    maze[y][x] = 0

    directions = [(0, -2), (2, 0), (0, 2), (-2, 0)]
    random.shuffle(directions)

    for dx, dy in directions:
        nx, ny = x + dx, y + dy

        if 1 <= nx < LARGEUR - 1 and 1 <= ny < HAUTEUR - 1:
            if maze[ny][nx] == 1:
                maze[y + dy // 2][x + dx // 2] = 0
                generate(nx, ny)

generate(1, 1)

# Entrée et sortie
start = (1, 1)
goal = (LARGEUR - 2, HAUTEUR - 2)

maze[start[1]][start[0]] = 0
maze[goal[1]][goal[0]] = 0
```

### Design du labyrinthe

```python
# Interface graphique
root = tk.Tk()
root.title("Labyrinthe")

canvas = tk.Canvas(
    root,
    width=LARGEUR * TAILLE_CASE,
    height=HAUTEUR * TAILLE_CASE
)
canvas.pack()

frame = tk.Frame(root)
frame.pack(pady=5)

btn_play = tk.Button(frame, text="▶ Play")
btn_pause = tk.Button(frame, text="⏸ Pause")
btn_restart = tk.Button(frame, text="🔄 Rejouer")

btn_play.pack(side=tk.LEFT, padx=5)
btn_pause.pack(side=tk.LEFT, padx=5)
btn_restart.pack(side=tk.LEFT, padx=5)

def draw_maze():
    for y in range(HAUTEUR):
        for x in range(LARGEUR):
            color = "black" if maze[y][x] == 1 else "white"

            canvas.create_rectangle(
                x * TAILLE_CASE,
                y * TAILLE_CASE,
                (x + 1) * TAILLE_CASE,
                (y + 1) * TAILLE_CASE,
                fill=color,
                outline=color
            )

    # Départ
    sx, sy = start
    canvas.create_rectangle(
        sx * TAILLE_CASE,
        sy * TAILLE_CASE,
        (sx + 1) * TAILLE_CASE,
        (sy + 1) * TAILLE_CASE,
        fill="green"
    )

    # Arrivée
    gx, gy = goal
    canvas.create_rectangle(
        gx * TAILLE_CASE,
        gy * TAILLE_CASE,
        (gx + 1) * TAILLE_CASE,
        (gy + 1) * TAILLE_CASE,
        fill="red"
    )

draw_maze()
```

### Résolution du labyrinthe

```python
visited = set()

steps = []

is_playing = False
current_step = 0

def solve(x, y):

    if (x, y) in visited:
        return False

    visited.add((x, y))

    if (x, y) != start and (x, y) != goal:
        steps.append(("visit", x, y))

    if (x, y) == goal:
        return True

    for dx, dy in [(0,-1),(1,0),(0,1),(-1,0)]:

        nx = x + dx
        ny = y + dy

        if (0 <= nx < LARGEUR and
            0 <= ny < HAUTEUR and
            maze[ny][nx] == 0):

            if solve(nx, ny):

                if (x, y) != start and (x, y) != goal:
                    steps.append(("path", x, y))

                return True

    return False
    
```

### Animation du labyrinthe

```python
def animate():

    global current_step

    if not is_playing:
        return

    if current_step >= len(steps):
        return

    action, x, y = steps[current_step]

    if action == "visit":

        canvas.create_rectangle(
            x * TAILLE_CASE,
            y * TAILLE_CASE,
            (x + 1) * TAILLE_CASE,
            (y + 1) * TAILLE_CASE,
            fill="deepskyblue"
        )

    elif action == "path":

        canvas.create_rectangle(
            x * TAILLE_CASE,
            y * TAILLE_CASE,
            (x + 1) * TAILLE_CASE,
            (y + 1) * TAILLE_CASE,
            fill="yellow"
        )

    current_step += 1

    root.after(20, animate)

def play():

    global is_playing

    if not is_playing:
        is_playing = True
        animate()


def pause():

    global is_playing

    is_playing = False


def restart():

    global current_step
    global is_playing

    current_step = 0
    is_playing = False

    draw_maze()

btn_play.config(command=play)
btn_pause.config(command=pause)
btn_restart.config(command=restart)

solve(*start)

root.mainloop()
```
